```python
complement = {"A":"T", "T":"A", "C":"G", "G":"C"}
```

```python
def reverse_complement(pattern):
    pattern_complement = [complement[nucliotide] for nucliotide in pattern]
    
    return ''.join(pattern_complement[::-1])
```
